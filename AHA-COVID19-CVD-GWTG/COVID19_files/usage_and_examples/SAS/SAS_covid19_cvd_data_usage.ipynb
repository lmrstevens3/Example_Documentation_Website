#### Creating a Library with SAS Code
The <code>LIBNAME</code> statement in the code below associates the name of the library, or libref, with the physical location of the library. <code>GWTG</code> is the name of the library or libref and <code>/mnt/workspace/GWTG/COVID19</code> is the physical location for the library.

```sas
LIBNAME GWTG "/mnt/workspace/GWTG/COVID19/data";
LIBNAME LIBRARY "/mnt/workspace/GWTG/COVID19/data";
```

**In SAS Studio,** Data can be imported by clicking the *New* button (far left button) under *Server Files and Folders* or selecting <code>Import Data under Tasks and Utilities > Utilities</code>.  
**Using SAS Code,** The following command is used to load datasets in SAS, where we set the path for the data. We use the library name <code>GWTG</code> in this case followed by a period(.) and then the SAS filename that you need to access the contents.
* <code>PROC IMPORT</code> can be used to load other files types, such as csv, if neeeded. 

```sas
DATA COVID_data;
    SET GWTG.covid19_cvd_aug20;
RUN;
```

### Viewing the data

The following commands are helpful when viewing large datasets. 
* <code>PROC CONTENTS</code> is used to list the contents present in the SAS files and to print the details for all the SAS files present in the directory. The <code>DETAILS</code> option also gives the information about the table variables.
  

```sas
PROC CONTENTS DATA=COVID_data DIRECTORY DETAILS;
RUN;
```

# <span style='width:50%; color:#C10E21; border-bottom:0.25rem  solid #f3f3f3;'>Tables and Summary Statistics</span>

### Tables and Frequencies 

Getting the counts of  the variables(or field names) can be done using the <code>PROC FREQ</code> method. This method is used for summarizing the categorical variables in the dataset.


```sas
PROC FREQ DATA= COVID_data; 
    TABLES METHDIAG/MISSING; /*This can be used with
                                any categorical variables in the dataset*/    
RUN;
```

### Summary Statistics  

The <code>PROC MEANS</code> computes the summary statistics for the dataset. This method can be used to compute the summary statistics for the continuous variables in the dataset.
    

```sas
PROC MEANS DATA=COVID_data MEAN MEDIAN STD MIN MAX Q1 Q3 NMISS MAXDEC=2 ; 
    VAR AGEI; /*This can be used for any continuous variable from the dataset. */
RUN;
```
